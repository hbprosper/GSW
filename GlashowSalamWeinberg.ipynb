## Glashow-Salam-Weinberg Model

### Introduction

In 1934, Enrico Fermi proposed the following 4-fermion model
\begin{align}
    \frac{G_F}{\sqrt{2}}(\overline{\psi}_e \gamma^\mu \psi_{\overline{\nu}})(\overline{\psi}_p \gamma_\mu \psi_n),
\end{align}
to describe $\beta$-decay processes such as $n \rightarrow p + e^- + \overline{\nu}_e$. The quantity $\overline{\psi} \equiv \psi^\dagger \gamma^0$ is called the Dirac conjugate of the spinor $\psi$, where the operator $^\dagger$ is the Hermitian conjugate, i.e., the transpose and complex conjugate and $\gamma^\mu$ are $4 \times 4$ matrices defined by the anti-commutation relations $\gamma^\mu \gamma^\nu + \gamma^\nu \gamma^\mu = 2  g^{\mu\nu}$, where $g^{\mu\nu}$ is the metric tensor of
special relativity.
$G_F$ is called the Fermi constant. 

Fermi assumed that $\beta$-decay involved vector currents, $\overline{\psi} \gamma^\mu \psi$, but in 1956 T. D. Lee and C. N. Yang noted that no experiment had been 
performed to test the assumption of parity invariance, that is, the invariance of particle interactions with respect to a reflection in a mirror. They suggested that certain experimental puzzles could be explained if parity was not conserved in the weak interactions. Furthermore, Lee and Yang suggested experiments to test this hypothesis. Shortly thereafter, Chien-Shiung Wu performed the experiment (see https://en.wikipedia.org/wiki/Wu_experiment) and confirmed the violation of parity invariance. The following year, Lee and Yang won the Nobel Prize in Physics. A year later, in 1958, Robert Marshak and his student George Sudarshan developed the vector axial-vector theory (V-A) of the weak interactions, which Richard Feynman and Murray Gell-Mann developed further.

We do not know why the weak interactions violate parity conservation, in particular, why they involve both vector $\overline{\psi} \gamma^\mu \psi$ and axial-vector $\overline{\psi} \gamma^\mu \gamma^5 \psi$ currents (
$\gamma^5 \equiv i \gamma^0 \gamma^1 \gamma^2 \gamma^3$). However, now that we have a better understanding of the weak interactions, it would be simpler to say that the weak interactions treat left and right-handed fields differently. But, the V-A jargon has stuck.


The Glashow-Salam-Weinberg (GSW) Langrangian (really a Lagrangian density) is built using the fields below for the 1st generation of leptons and quarks:

\begin{align}
L & = \begin{pmatrix} \nu_L \\\\ e_L \end{pmatrix},\quad
e_R, \\
Q & = \begin{pmatrix} u_L \\\\ d_L \end{pmatrix},\quad
u_R, \quad d_R, 
\end{align}
where
\begin{align}
P_L & = \frac{1}{2}(1 - \gamma^5),\\
P_R & = \frac{1}{2}(1 + \gamma^5),
\end{align}
are projection operators that split a fermion field into its left and right components,
$\psi_{L,\, R} = P_{L,\, R}  \, \psi$
such that $\psi = \psi_L + \psi_R$.
Note the absence, in this model, of the right-handed neutrino field $\nu_R$.

The GSW model, inspired by the highly successful quantum field theory called quantum electrodynamics (QED) and the quantum field theories developed by Yang and Mills, is defined by the Langrangian,
\begin{align}
    {\cal L} & = \overline{L} i \gamma^{\mu} ( \mathbf{1} \partial_{\mu} + i g_1 y^L_L \mathbf{1} B_{\mu} + i g_2 \mathbf{T} \cdot \mathbf{W}_{\mu} ) L \nonumber\\ 
    & + \overline{Q} i \gamma^\mu (\mathbf{1}\partial_\mu + i g_1 y_L^Q \mathbf{1} B_\mu  + i g_2 \mathbf{T} \cdot \mathbf{W}_\mu ) Q\nonumber\\
    & + \overline{e}_R i \gamma^\mu ( \partial_\mu  + i g_1 y^e_R B_\mu  ) e_R\nonumber\\
    & + \overline{u}_R i \gamma^\mu ( \partial_\mu + i  g_1 y^u_R B_\mu  ) u_R\nonumber\\
    & + \overline{d}_R i \gamma^\mu ( \partial_\mu + i g_1 y^d_R B_\mu  ) d_R \nonumber \\
    & - \frac{1}{4} \mathbf{W}_{\mu\nu} \cdot \mathbf{W}^{\mu\nu}\nonumber\\
    & - \frac{1}{4} B_{\mu\nu} B^{\mu\nu},
\end{align}

which explicitly treats left-handed and right-handed fields differently; that is, the model explicitly violates parity invariance. The symbols, $y$, which related to the constants $Y = 2 y$ called *hypercharges*, are chosen to obtain agreement with observations. We use $y$ instead of $Y / 2$, as is the usual convention, to make the expressions a bit tidier. The
other symbols are:
\begin{align*}
\mathbf{1} & \textrm{ a $2 \times 2$ unit matrix}, \\
    \mathbf{T} & = \boldsymbol{\tau / 2} \textrm{ where $\tau$ are the $2 \times 2$ Pauli matrices} \nonumber\\ 
    & \, \textrm{and the tensors $W_{\mu\nu}$ and $B_{\mu\nu}$ are given by}\nonumber\\
    \mathbf{W}_{\mu\nu} & = \partial_\mu \mathbf{W}_\nu- \partial_\nu \mathbf{W}_\mu - g_2 \, \mathbf{W}_\mu \times \mathbf{W}_\nu,\\
    (\mathbf{W}_{\mu\nu})_i   & = (\partial_\mu \mathbf{W}_\nu)_i - (\partial_\nu \mathbf{W}_\mu)_i - g_2 \, \epsilon_{ijk} W^j_\mu W^k_\nu, \\
    B_{\mu\nu} & = \partial_\mu B_\nu - \partial_\nu B_\mu \textrm{ and }\\
    \mathbf{W}_\mu & = (W^1_\mu, \, W^2_\mu, \, W^3_\mu).
\end{align*}
The quantities enclosed in the first and second pair of parentheses in the Lagrangian are $2 \times 2$ matrices that act on the lepton and quark doublet fields $L$
and $Q$, respectively, while the parentheses in the remaining fermionic terms enclose $1 \times 1$ matrices, that is, scalars, which act on the singlet fields $e_R$, $u_R$, and $d_R$, respectively. But note, each field is a 4-component Lorentz object: a 4-component Dirac spinor in the case of the fermionic fields and a 4-vector in the case of the bosonic fields.

In the GSW Lagrangian, because the hypercharges are assigned specific numerical values the Lagrangian has only two free parameters, namely, the couplings $g_1$ and $g_2$.

### $SU(2) \otimes U(1)$ Gauge Invariance
The GSW Lagrangian is invariant under
the group $SU(2) \otimes U(1)$. The elements $U_2$ of  $SU(2)$ can be represented as the $2 \times 2$ matrices

\begin{align*}
    U & = \exp(i \mathbf{T} \cdot \boldsymbol{\theta}(x)), \textrm{ while}\\
    U & = \exp(i y \, \alpha(x) ),
\end{align*}

represent the elements of $U(1)$. The $SU(2)$ elements are defined by three spacetime-dependent functions, $\mathbf{\theta}(x) = (\theta_1(x)$, $\theta_2(x)$, $\theta_3(x))$, each associated with group generators $T_1 = \tau_1 / 2$, $T_2 = \tau_2 / 2$, and $T_3 = \tau_3 /  2$, respectively, while the
$U(1)$ elements are defined by a single function $\alpha(x)$ and a single generator $y = Y \, / \, 2$.

Invariance under the group $SU(2) \otimes U(1)$ means that the following replacements
leave the Lagrangian unchanged,
\begin{align*}
U \in \boldsymbol{SU(2)} \quad \quad \\
L & \rightarrow U L, \quad Q \rightarrow U Q, \\
e_R & \rightarrow e_R, \quad u_R \rightarrow u_R, \quad d_R \rightarrow d_R \\
 \mathbf{T} \cdot \mathbf{W}_\mu & \rightarrow U ( \mathbf{T} \cdot \mathbf{W}_\mu  ) U^{-1} - \frac{1}{i g_2} (\partial_\mu U) U^{-1}, \\
B_\mu & \rightarrow B_\mu, \\ 
U \in \boldsymbol{U(1)} \quad \quad \\
L & \rightarrow U L, \quad Q \rightarrow U Q \\
e_R & \rightarrow U e_R, \quad u_R \rightarrow U u_R, \quad
d_R \rightarrow U d_R, \\
\mathbf{W}_\mu & \rightarrow \mathbf{W}_\mu, \\
B_\mu & \rightarrow U B_\mu U^{-1} - \frac{1}{i  g_1} (\partial_\mu U) U^{-1}.
\end{align*}

Notice the absence of mass terms such as $m \psi \psi$ and $M_B^2 B_\mu B^\mu$ in the Lagrangian. This is necessary because explicit mass terms violate the $SU(2) \otimes U(1)$ invariance. 

The model above was developed by Glashow in 1961. However, because it predicted zero mass for all bosonic fields (not just for the photon), it clearly did not accord with the absence of experimental evidence for a long-range weak nuclear force. But in 1964 a mechanism for introducing mass without violating gauge invariance was developed independently by Englert & Brout, by Higgs, and by Guralnik, Hagen & Kibble in a series of papers in that year. These papers were inspired by the work of the condensed matter theorist Phil Anderson. Much to the chagrin of the other theorists, the mechanism was referred to, by Steven Weinberg, as the Higgs mechanism and the name stuck. A more accurate name, though a bit of a mouthful, would be the __Anderson-Englert-Brout-Higgs-Guralnik-Hagen-Kibble__ mechanism. 

In 1967, Salam and Weinberg independently showed how the Higgs mechanism can be used to correct the mass problem of the Glashow model. In addition, Weinberg introduced gauge invariant (Yukawa) interactions between the Higgs field and the fermion fields that induced mass terms for the latter. Thus was born the __electroweak model__. 

It is usually argued that __local gauge__ invariance means that the phases of the quantum fields can be set to different values at different events in spacetime and that this makes sense because there is no reason why, for example, Andromedans, ten million years from now, should adopt the same phase convention as Earthlings do now. Personally, I find this argument unconvincing. There exist, after all, other symmetries with phases that remain constant over the whole of spacetime. For example, the replacements (see
https://arxiv.org/pdf/hep-ph/0410370.pdf, p. 9)
\begin{align*}
    Q & \rightarrow e^{i \theta / 3} Q,\\
    u_R & \rightarrow e^{i \theta / 3} u_R, \\
    d_R & \rightarrow e^{i \theta / 3} d_R, \\
    \textrm{and} & \\
    L &  \rightarrow e^{i \phi } L, \\
    e_R & \rightarrow e^{i \phi } e_R,
\end{align*}

which correspond to baryon number and lepton number conservation, respectively, are __global symmetries__ that leave the Lagrangian invariant. Admittedly, these global symmetries were not imposed but are inherent in the structure of the theory and are,
in that sense, *accidental*. My question is:
why is it fine to have global symmetries where certain quantities, such as $\theta$ and $\phi$ are the same everywhere and everywhen, that is, Andromedans and Earthlings must use the same values of these phases even though they have never met and yet it is not
fine to have the same gauge field phases everywhere? There is no good reason that I can discern. I think it is better simply to accept the 
discovery that if one imposes local gauge invariance we are led to theories, which for reasons not 
yet clear, are spectacularly successful. 

### Hypercharges
We need a name for $y \equiv Y / 2$; let's refer to $y$ as the *reduced* hypercharge. The reduced hypercharge $y$, electric charge $Q$ (in units of the proton charge and not to be confused with the quark doublet), and the third component of the (weak) isospin $I_3 = \pm 1/2$ are related by the Gell-Mann Nishijima relation $y = Q - I_3$.
For the left-handed neutrino field, $Q = 0$ and $I_3 = +1/2$, therefore, the reduced hypercharge is $y^L_L = -1/2$.
For the left-handed electron field, $Q = -1$ and $I_3 = -1/2$ and again $y^L_L = -1/2$. On
the other hand, since the right-handed electron field is an $SU(2)$ singlet $I_3 = 0$ and, therefore, $y^e_R = -1$. We conclude that the reduced hypercharge is a property of the doublet or singlet. 

In the quark model, up quarks have electric charge $Q = +2/3$ (in units of the proton charge), while down quarks have electric charge $Q =  -1/3$. 
Therefore, in order to satisfy the Gell-Mann Nishijima relation we must make the
assignments $y^Q_L = 1/6$, $y^u_R = 2/3$, and $y^d_R = -1/3$. The hypercharge assignments are collected below.

| Reduced Hypercharge      | Value |
| :---------- | ----------- |
| $y^Q_L$ | $+1/6$ |
| $y^L_L$ | $-1/2$       |
| $y^u_R$ | $+2/3$ |
| $y^d_R$ | $-1/3$|
| $y^e_R$ | $-1$        |

### Bosons
If we define $X_\mu \equiv \mathbf{T} \cdot \mathbf{W}_\mu$ and $X_{\mu\nu} \equiv \mathbf{T} \cdot\mathbf{W}_{\mu\nu}$, and noting that $T^a T^b = i \frac{1}{2}\epsilon_{abc} T^c + \frac{1}{4} \delta_{ab}\mathbf{1}$ and,
therefore, $\textrm{Tr}(T^a T^b) = \frac{1}{2} \delta_{ab}$, we can write
\begin{align*}
    X_{\mu\nu} & = \partial_\mu X_\nu - \partial_\nu X_\mu + i g_2 [X_\mu, X_\nu] ,
\end{align*}
and, therefore,
\begin{align*}
\mathbf{W}_{\mu\nu} \cdot \mathbf{W}^{\mu\nu} & = \frac{1}{2}\textrm{Tr}(X_{\mu\nu} X^{\mu\nu}) .
\end{align*}

In the low energy limit ($q^2 \ll M_W$, where the W boson mass $M_W \sim 80\,$GeV), the GSW model reproduces the Fermi model for $\beta$-decay provided that we set
$$\frac{G_F}{\sqrt{2}} = \frac{g_2^2}{8 M_W^2}.$$

## An Aside
The Lagrangian, when integrated over spacetime 
$$S = \int d^4x {\cal L},$$
yields the action, which, in units in which $c = \hbar = 1$, is a dimensionless quantity. In these units, the measure $d^4x$ has dimensions $[M]^{-4}$, where $M$ is mass. Therefore, necessarily, every term in the Lagrangian must have dimension $[M]^4$. Consider a quantum field theory that is a perturbed version of the Standard Model. We can write such a model, schematically, as follows,

$${ \cal L} = {\cal L}_\text{SM} + \sum_{n} \frac{c_n}{\Lambda^{n-4}} O_n,$$

where $\Lambda$ is a scale with dimension $[M]$, $c_n$ are called Wilson coefficients,  and $O_n$ are operators of dimension $[M]^n$. The form above is an example of an **effective field theory** (EFT). The idea is that the Lagrangian is a low-energy vestige of a deeper, unknown, theory.

## Code to display long expressions

```python
def display_grouped(expr, terms_per_line=2):
    from sympy import Add, latex
    from IPython.display import display, Math

    if type(expr) == (type([]) or type(())):
        terms = expr
    else:
        terms = Add.make_args(expr)
        
    lines = []
    for i in range(0, len(terms), terms_per_line):
        chunk = terms[i:i+terms_per_line]
        lines.append('+'.join(latex(t) for t in chunk))
        
    latex_str = r'\begin{aligned}' + '\n'
    latex_str += r' \\'.join(f'& {line}' for line in lines)
    latex_str += '\n' + r'\end{aligned}'
    latex_str = latex_str.replace('+-', '-') # A hack!

    display(Math(latex_str))

def sort_by_gauge_field(expr, gauge_fields):
    """
    Sort terms in a sum by gauge field, grouping terms with the 
    same gauge field together.
    gauge_fields: ordered list of gauge field symbols, e.g. [B_mu, Wp_mu, Wm_mu, W3_mu]
    """
    from sympy import Add
    
    def gauge_key(term):
        for i, field in enumerate(gauge_fields):
            if field in term.free_symbols:
                return i
        return len(gauge_fields)  # unrecognised fields go last
    
    terms = Add.make_args(expr)
    sorted_terms = sorted(terms, key=gauge_key)
    return sorted_terms

# Usage
# gauge_fields = [B_mu, Wp_mu, Wm_mu, W3_mu]
# sorted_terms = sort_by_gauge_field(expr, gauge_fields)
# display_grouped(sorted_terms, terms_per_line=2)
```

## The unification of the electromagnetic and weak nuclear forces 
The 1961 Glashow hypothesis is that the electromagnetic and weak nuclear forces are  unified in the following sense:

\begin{align}
    \begin{pmatrix} B_\mu \\ W^3_\mu \end{pmatrix} & = 
    \begin{pmatrix} 
        \cos\theta_W & -\sin\theta_W \\
        \sin\theta_W & \cos\theta_W
    \end{pmatrix}     \begin{pmatrix} A_\mu \\ Z_\mu \end{pmatrix} .
\end{align}

Above a certain critical temperature, the symmetry between the two forces (called the electroweak symmetry) is restored and the two forces merge into a single __electroweak force__. Below the critical temperature, the electroweak symmetry is said to be *spontaneously broken* by the change in shape of the potential energy of the Higgs field, which we discuss below. Because of the changed shape (which, to date, is introduced by hand), the theory yields infinitely many vacuum state solutions. While this infinite *ensemble* of vacuum state solutions collectively respect the electroweak symmetry, the particular vacuum state into which the universe settles does not. 

This feature is analogous to that of an infinite ferromagnet. Above a certain critical temperature, called the Curie temperature, the magnetic moments of the ferromagnet's domains are randomly oriented. Therefore, when averaged over the whole ferromagnet there is no net magnetization and the ferromagnet exhibits an $O(3)$ symmetry, that is, it is invariant with respect to rotations in 3-D space. However, well below the Curie temperatue, the magnetic moments line up in the same randomly selected direction, which causes the system to lose its $O(3)$ symmetry. Thus, the ground state of the ferromagnetic with its net magnetization does not share the symmetry of the equations that describe the system. However, collectively, the infinite ensemble of possible ground states still preserves the symmetry in the sense that a simultaneous rotation of all of the ground states in the same way doesn't change the ensemble. 

Broken symmetry is our friend because without it the universe would be far less diverse and we would not exist.

In lines 2 and 3, in the expression above, there are interactions between the electromagnetic field $A_\mu$ and the neutrino current $\bar{\nu}_L \gamma^\mu \nu_L$. However, there is no evidence that such interactions exists. So Glashow did the simplest possible thing: he assumed that  $$-y^L_L \, g_1\cos(\theta_W) - \frac{g_2}{2}\sin(\theta_W) = 0.$$ 

```python
a = a.subs(-YLL*g1*sm.cos(thetaW), g2*sm.sin(thetaW)/2)

sorted_a = sort_by_gauge_field(a, gauge_fields)
display_grouped(sorted_a, terms_per_line=2)
```

So far, so good! The non-existent interactions are gone. Now, if this theory has anything to do with reality, we need to recover the known form of the electromagnetic interactions, which for electrons has the form $$q  A_\mu (\bar{e}_L \gamma^\mu e_L + \bar{e}_R \gamma^\mu e_R ),$$ where $q$ is the magnitude of the electic charge of the electron. We can recover the correct form of the QED interactions
if we impose the conditions

\begin{align*}
    -y^e_R g_1 \cos\theta_W & = g_2 \sin\theta_W = q,
\end{align*}

which, recalling that $y^e_R = -1$, yields 

\begin{align*}
    \tan\theta_W & = \frac{g_1}{g_2},\\
    \text{and} &,\\
    q & = \frac{g_1 g_2}{\sqrt{g_1^2 + g_2^2}}.
\end{align*}
Already we see a few beautiful consequences of Glashow's inspired work, namely, the link between the electric charge and the couplings in the Lagrangian. Let's now write the interactions in terms of $q$.

```python
a = a.subs({-YeR*g1*sm.cos(thetaW): q, g2*sm.sin(thetaW): q })
a = a.subs({g1*sm.cos(thetaW): q})
a = a.subs({g1 : q / sm.cos(thetaW), g2: q / sm.sin(thetaW)})
a = a.trigsimp().expand()

display_grouped(a, 2)
```

### Currents associated with $W^+_\mu$
We extract the coefficients of the bosonic fields, that is, the currents, and manipulate them into the forms in which they are typically expressed. 

```python
a = a.collect(Wplus)
Wpluscoeff  = a.coeff(Wplus)
Wpluscoeff  = Wpluscoeff.subs(g2, q/sm.sin(thetaW)).factor(q).simplify()
Wpluscoeff
```

### Currents associated with $W^-_\mu$

```python
a = a.collect(Wminus)
Wminuscoeff = a.coeff(Wminus)
Wminuscoeff = Wminuscoeff.subs(g2, q/sm.sin(thetaW)).factor(q).simplify()
Wminuscoeff
```

### Currents associated with $A_\mu$

`sympy` is unable to collect noncommutative variables, so we hack it! Replace noncommutative variables by commutative ones, collect, and go back to the original variables.

```python
N, E, U, D = sm.symbols('N, E, U, D')

a = a.collect(A)

Acoeff = a.coeff(A).expand()

Acoeff = Acoeff.subs({ebarL*gamma*eL: E,
                      ubarL*gamma*uL: U, 
                      dbarL*gamma*dL: D})
Acoeff = Acoeff.collect(E).collect(U).collect(D)

Acoeff = Acoeff.subs({E:ebarL*gamma*eL, 
                      U: ubarL*gamma*uL, 
                      D: dbarL*gamma*dL})

Acoeff = Acoeff.subs(g1, g2*sm.sin(thetaW)/sm.cos(thetaW))
Acoeff = Acoeff.subs({g2*sm.sin(thetaW): q})

display_grouped(Acoeff, 4)
```

Notice that the coefficients of the left-handed quark currents can be written as $-q(y^Q_L \pm 1/2)$, that is, $-q(y^Q_L + I^f_3) = -q Q^f$, where $f = u, d$. Also, since the 3rd component of weak isospin for singlets is $I_3 = 0$, we have $y^f_R = Q^f$.

```python
Acoeff = Acoeff.subs({-q*YQL + q/2: -q*Qd,
                      -q*YQL - q/2: -q*Qu, 
                      YuR: Qu, 
                      YdR: Qd})
display_grouped(Acoeff, 2)
```

Likewise, we notice that the coefficients of the electron currents can be written as $-Q^e q$:

```python
# left electron current
Acoeff = Acoeff.subs({q*ebarL*gamma*eL: -Qe*q*ebarL*gamma*eL})

# right electron current
Acoeff = Acoeff.subs({q*ebarR*gamma*eR: -Qe*q*ebarR*gamma*eR})

display_grouped(Acoeff, 2)
```

..which, collecting terms with the same charge yields:

```python
Acoeff = Acoeff.\
collect(-Qd*q).\
collect(-Qu*q).\
collect(-Qe*q)

display_grouped(Acoeff, 1)
```

We see that the electromagnetic current $J^\mu_\textrm{em}$ can be written as
\begin{align*}
J^\mu_\textrm{em} & = - q \sum_{f = u, d, e} Q^f \overline{f} \gamma^\mu f,
\end{align*}
where $f = f_L + f_R$, noting that $P_{L,R} \gamma^\mu P_{L, R} = 0$, an $Qd^f$ is the electic charge in units of the proton charge for the quantum field of flavor $f$. Notice, also, in accordance with experiment, the electromagnetic interactions do not violate parity invariance: the coupling to the left-handed and right-handed fields are the same.

### Currents associated with $Z_\mu$

As noted above, collect does not work with noncommutative symbols. So, again, use our hack.

```python
a = a.collect(Z)

Zcoeff = a.coeff(Z)

Zcoeff = Zcoeff.subs({nubarL*gamma*nuL: N, 
                      ebarL*gamma*eL: E,
                      ubarL*gamma*uL: U, 
                      dbarL*gamma*dL: D})
Zcoeff = Zcoeff.collect(E).collect(N).collect(U).collect(D)

Zcoeff = Zcoeff.subs({N:nubarL*gamma*nuL, 
                      E:ebarL*gamma*eL, 
                      U: ubarL*gamma*uL, 
                      D: dbarL*gamma*dL})
Zcoeff = Zcoeff.subs(g1, g2*sm.sin(thetaW)/sm.cos(thetaW))

Zcoeff = Zcoeff.trigsimp()

display_grouped(Zcoeff, terms_per_line=1)
```

### Weak neutral currents
The weak neutral current $J^\mu_\textrm{nc}$ can be written as
\begin{align*}
J^\mu_\textrm{nc} & = q \tan(\theta_W) \sum_{f=d, u, e} y^f_R \overline{f}_R \gamma^\mu f_R \\
& + q \tan(\theta_W)\sum_{f=d, u, e, \nu} \left(y^f_L  - \frac{I^f_3}{2 \tan^2(\theta_W)} \right) \overline{f}_L \gamma^\mu f_L, 
\end{align*}

which predicts the existence of neutral currents that violate parity in a specific way.

The more traditional way of writing the weak neutral current $J^\mu_\textrm{nc}$ is
\begin{align*}
J^\mu_\textrm{nc} & = \frac{2q}{\sin(2\theta_W)} \left(

& +
\frac{2q}{\sin(2\theta_W)} \left( \sum_{f = d, u, e, \nu} (-I^f_3 + Q^f \sin^2\theta_W) \, \overline{f}_L \gamma^\mu f_L   \right).
\end{align*}

## The Higgs Field
The Lagrangian of the Higgs field is given by
\begin{align}
    {\cal L}_\textrm{Higgs} &= (D^\mu \phi)^\dagger (D_\mu \phi) - V(\phi^\dagger \phi) ,
\end{align}
where
\begin{align}
    D_\mu &= \partial_\mu + 
    i g_1 y^\phi_L B_\mu + i g_2 \mathbf{T} \cdot \mathbf{W}_\mu,\\
    V(\phi^\dagger \phi) &= \mu^2 \phi^\dagger \phi + \lambda (\phi^\dagger \phi)^2, \quad\textrm{and the 4-component Higgs field is given by}\\
    \phi & = \begin{pmatrix}
    \phi_1 + i \phi_2 \\
    \phi_3 + i \phi_4
    \end{pmatrix} .
\end{align}

#### Qualitative description of the Higgs mechanism
The stability of the vacuum requires $\lambda > 0$. In the very early universe, we presume that $\mu^2 > 0$, in which case $\langle 0| \phi | 0 \rangle = 0$, that is, the expectation (i.e., average) value of the Higgs field in the vacuum is zero. But, at some critical temperature, $T_\textrm{crit}$, there is a phase transition in which it is presumed that $\mu^2$ changes sign thereby causing the Higgs potential to develop infinitely many degenerate vacua in each of which the Higgs field is non-zero. Above $T_\textrm{crit}$, there is a unique vacuum state and this state respects the symmetry of the Lagrangian. Below $T_\textrm{crit}$, the *ensemble* of vacua respects the symmetry, but, the universe cools into one of the infinitely many potential vacua, which on its own breaks the symmetry. This is an example of *spontaneous symmetry breaking*, which we touched upon above. The specific instance is called electroweak symmetry breaking (EWSB).  

The vacuum state of the universe is one in which the Higgs field behaves like a superconductor that attenuates the propagation of the $W_\mu$ and $Z_\mu$ fields, thereby rendering these fields massive, but permits the free propagation of the $A_\mu$ field. (The theory is, of course, made to do this!) Moreover, the interactions between the Higgs field and the fermions, which also are put in by hand, cause a rapid flipping between the left-handed and right-handed fermion fields, which causes energy to build up within them that we interpret as the mass of the associated field quanta, that is, particles. 
The results from the LHC experiment are, so far, consistent with this picture.

Since there is no right-handed neutrino the flipping mechanism cannot occur and therefore
the neutrino remains massless in the GSW model. However, neutrinos are now known to have (very small) masses,  but, how their masses arise remains an open question. 

#### Quantum numbers
We need to specify the value of the reduced hypercharge $y^\phi_L$ associated with the doublet $\phi$. As above,  the reduced hypercharge is chosen so that the Gell-Mann Nishijima relation $y = Q - I_3$ holds, where the third component of the weak isospin $I_3 = +1/2$ for the upper component of the doublet and $I_3 = -1/2$ for the lower component. In the GSW model, it is assumed that the vacuum expectation of $\phi$ is
$$\langle 0|\phi|0 \rangle = \frac{1}{\sqrt{2}}\begin{pmatrix} 0 \\ v \end{pmatrix}.$$ Since the vacuum has a net electric charge of zero, we must set $Q = 0$ for the lower component $\phi_3 + i \phi_4$. From $y^\phi_L = Q - I_3$ it follows that $y^\phi_L = +1/2$. This also shows that the upper component, $\phi_1 + i \phi_2$ must be assigned $Q = +1$, which of course is why it is assumed to be zero in the vacuum otherwise the entire universe would have a huge net positive electric charge!

#### Higgs field at low energies
At energies not too far from the vacuum state (which is where we live!), it is assumed that the Higgs field, $\phi$, can be approximated as follows
$$\phi  = \frac{1}{\sqrt{2}}\begin{pmatrix} 0 \\ v + H\end{pmatrix},$$
where $H$ represents the fluctuations of the Higgs field about its 
average value $v \approx 246 \,\textrm{GeV}$ in the vacuum state (the vacuum expectation value). The quanta of the field $H$ are identified with the Higgs boson discovered at CERN in 2012. There may be absolutely no significance whatsoever to the numerical fact that $v / \sqrt{2} \approx m_t$, where $m_t$ is the mass of the top quark; or it could be deeply significant. Whatever conclusion is to be drawn, the value of $v$ implies (as we shall see below) that the coupling of the top quark to the Higgs boson is close to (and may even be) unity.   

In the following, we shall also need the conjugate field defined by

$$\phi_c \equiv i \tau_2 \phi = \frac{1}{\sqrt{2}}\begin{pmatrix} v + H \\ 0 \end{pmatrix}.$$

### Vector boson - Higgs boson interactions

In order to see explicitly what interactions are implied by the Higgs field Langrangian, we first compute
\begin{align}
    F &= g_1 y^\phi_L B_\mu +  g_2 \mathbf{T} \cdot \mathbf{W}_\mu,
\end{align}
which appears in the covariant derivative $D_\mu = \partial_\mu + 
    i g_1 y^\phi_L B_\mu + i g_2 \mathbf{T} \cdot \mathbf{W}_\mu$.

#### Note
The matrix $F$ is Hermitian, that is, $F^\dagger = F$.


```python
v   = sm.symbols('v')
H   = sm.symbols('H')
phi = Matrix([0, (v+H)/sm.sqrt(2)])
F   = g1*YphiL*b + g2*w
phic= sm.I*tau2*phi
F, phi, phic
```

Then, compute

\begin{align}
    x & = F \phi, \\
    \overline{x} & = \phi^T F^\dagger,\\
            & = \phi^T F,\\
            f & = \overline{x} x.
\end{align}

```python
x    = F*phi
xbar = phi.T*F
f = xbar*x
f = f[0]
f = f.subs({YphiL: 1})
f = f.subs({Wminus: Wminusup})
f
```

Map to $Z$ field.

```python
f1 = f.subs(g1*B - g2*W3/2, Z * sm.sqrt(g1**2 + g2**2)/2)
f1 = f1.expand()
f1 = f1.subs({Z**2: Z*Zup})
f1 = f1.collect(Z*Zup*v**2).collect(Z*Zup*H*v).collect(Z*Zup*H**2)
f1 = f1.simplify()

sorted_f1 = sort_by_gauge_field(f1, [H]+gauge_fields)
display_grouped(sorted_f1, terms_per_line=3)
```

Collect terms

```python
f2 = sm.Add(*sorted_f1).collect(H**2*Zup*Z/8).collect(H*Zup*Z*v/4)
sorted_f2 = sort_by_gauge_field(f2, [H])
display_grouped(sorted_f2, 2)
```

The terms involving the Higgs boson field $H$ in the low-energy limit are the single and di-Higgs boson interactions, while the terms $\propto v^2$ are mass terms. For a massive vector boson field, $V_\mu,$ the mass term has the form $(m_V^2 \, / \, 2) V_\mu V^\mu$. The term $W^+_\mu W^{- \mu}$ is the sum of two terms quadratic in the fields $W^1$ and $W^2$, with the same mass, so the mass term has the form $m_W^2 \, W^+_\mu W^{- \mu}$. Therefore, in the GSW model the $W$ boson mass is predicted to be $m_W = v g_2 \, / \, 2$, while for the $Z$ boson the prediction is $m_Z = v \sqrt{g_1^2+g_2^2} \, / \, 2$ and, therefore, $m_W / m_Z = g_2 / \sqrt{g_1^2 + g_2^2} = \cos\theta_W$, which modulo higher order corrections is in excellent agreement with measurements as shown below.

```python
sw2 = 0.2397             # sin^2(theta_W)
sw  = sm.sqrt(sw2)
cw  = sm.sqrt(1-sw2)
print('cos(theta_W) = %8.3f' % cw)
```

```python
MW  = 80.385 # GeV
MZ  = 91.188 # GeV
cwp = MW/MZ
print('M_W / M_Z    = %8.3f\t%8.3f' % (cwp, cw / cwp))
```

We see that the couplings, $g_{HWW}$ and $g_{HZZ}$, of the Higgs field to the vector bosons are predicted to be $2 m_W^2/v$ and $m_Z^2 / v$, respectively. The quartic couplings, $g_{HHWW}$ and $g_{HHZZ}$, of two Higgs fields to two vector bosons are predicted to be considerably smaller, $(m_W / v)^2$ and $\frac{1}{2}(m_Z / v)^2$, respectively.

### Yukawa interactions between fermions and the Higgs field

The interactions between the fermions and the Higgs field are presumed to be Yukawa interactions (Weinberg) of the form

\begin{align*}
-\sqrt{2} g_e (\overline{L} \phi e_R + \overline{e}_R \phi^\dagger L)
-\sqrt{2} g_u (\overline{Q} \phi_c u_R + \overline{u}_R \phi_c^\dagger Q)
-\sqrt{2} g_d (\overline{Q} \phi d_R + \overline{d}_R \phi^\dagger Q)
\end{align*}

Let us pause for a minute to consider the dimensions of the above interactions. In natural units, the Lagrangian has units of $[M]^4$ since the measure over spacetime, $d^4x$, has dimensions of $[M]^{-4}$. Since a fermion mass term looks like 

$$m \overline{\psi} \psi$$

and $\textrm{dim}(m) = [M]$, it follows that $\overline{\psi} \psi$ must have dimension $[M]^3$ and, therefore, a fermion field must have dimension $[M]^{3/2}$.  The Higgs boson has dimension $[M]$. Consequently, the Yukawa interactions between fermions and the Higgs boson must be a gauge invariant combination of *two* fermion fields with a combined dimension of $[M]^3$ and the Higgs field with dimension $[M]$. Moreover, since the Higgs field is a doublet, it must be paired with the Dirac conjugate of a *doublet* fermion field, $L$, giving a term of dimension $[M]^{5/2}$. When this term is paired with a *singlet* fermion field, $f_R$, of dimension $[M]^{3/2}$ we arrive at terms, e.g., $\overline{L} \phi e_R$ and its Hermitian conjugate $\overline{e}_R \phi^\dagger L$, of the correct dimension.  However, there is a subtlety. In order to avoid flavor changing Yukawa interactions, we need to fudge the interaction of the up quark field with the Higgs field by using the conjugate field $H_c$ rather than $H$. 

Define coupling parameters and Higgs conjugate fields

```python
ge, me = sm.symbols('g_e, m_e')
gu, mu, gd, md = sm.symbols('g_u, m_u, g_d, m_d')
gu, mu, gd, md
ge, me, gu, mu, gd, md
```

```python
phi, phic
```

Expand 

$-\sqrt{2} g_e (\overline{L} \phi e_R + \overline{e}_R \phi^\dagger L)$

```python
Le = -ge*(Lbar*phi*eR + ebarR*phi.T*L)[0]*sm.sqrt(2)
Le = Le.expand()
Le = Le.collect(-ge*v).collect(-ge*H)
Le = Le.subs(ge*v, me)
Le = Le.subs(ge, me/v)

Le
```

Let's now do the same thing for the quarks.

```python
Lq = -gd*(Qbar*phi*dR  + dbarR*phi.T*Q)[0]*sm.sqrt(2) \
     -gu*(Qbar*phic*uR + ubarR*phic.T*Q)[0]*sm.sqrt(2)
Lq = Lq.expand()
Lq = Lq.collect(-gu*v).collect(-gu*H)
Lq = Lq.collect(-gd*v).collect(-gd*H)
Lq = Lq.subs(gu*v, mu)
Lq = Lq.subs(gd*v, md)
Lq = Lq.subs(gu, mu/v)
Lq = Lq.subs(gd, md/v)

Lq
```

We see that the Yukawa couplings, together with the particular choice of the value of the Higgs field in the vacuum state, yields two important consequences as alluded to above: the first is mass terms for the fermions and the second is interactions between the Higgs boson and fermions that are proportional to the fermion mass and inversely proportional to the vacuum expectation value $v$. This implies that the top quark coupling to the Higgs boson is close to, and may be, unity. Why that is, is yet another mystery in a long list of them. 

We conclude that the Higgs mechanism generates mass terms for the weak vector bosons as well as the fermions while preserving gauge invariance.

### The Cabibbo Hypothesis
So far we have considered only the 1st generation of particles. But three generations have been discovered so far and we have implicitly assumed that there is no "crosstalk" between them. There are, however, decays that  can be explained if "crosstalk" exists between generations. For example, the existence of decays such as $$\Lambda (d, u, s) \rightarrow p (d, u, u) \, \, e^- \, \, \bar{\nu}_e$$ are readily explained by supposing that the $W$ boson can convert an s-quark to a u-quark, that is, a 2nd generation quark to a 1st generation quark. Further clues are provided by the rates of decays such as
\begin{align}
    \pi^+ & \rightarrow \mu^+ \, \nu_\mu,\\
    K^+ & \rightarrow \mu^+ \, \nu_\mu,
\end{align}
which are measured to be in the ratio
\begin{align}
    \frac{\Gamma(\pi^+ \rightarrow \mu^+ \, \nu_\mu)}
         {\Gamma(K^+ \rightarrow \mu^+ \, \nu_\mu)} & \approx \frac{1}{20}.
\end{align}
The above decays can be described with the following Feynman diagrams,
![decays](decays01.png)
in which the couplings $g_2\textrm{?}$ differ from $g_2$ by a factor chosen to match the measured ratio. The pictorial explanation of these decays, just like the decay of positronium (an electron-positron bound system), is that the particles comprising the $\pi^+$ annihilate, in this case to a W boson that subsequently decays leptonically. The observed ratio is readily explained if the W boson can interact with quarks of differing generations, but at a reduced rate compared with that of within-generation interactions. These observations and many others were explained by the Italian physicist Cabibbo who suggested that the lower component of the quark doublet, that is, the down quark field, be replaced with the combination
$$d_L \rightarrow d_L^\prime = \cos\theta_C \, d_L + \sin\theta_C \, s_L,$$
where $\theta_C$ is called the Cabibbo angle. With a suitable choice of the Cabibbo angle (about 12$^\textrm{o}$), the Cabibbo hypothesis worked well. 
Unfortunately, however, the replacement
\begin{align}
 \overline{d}_L \gamma^{\mu} d_{L} \rightarrow \overline{d^\prime}_L \gamma^{\mu} d^\prime_{L}, 
\end{align}
predicts the existence of flavor-changing neutral currents (FCNC) at a rate in sharp disagreement with observations. 

```python
thetaC = sm.symbols('theta_C')
xbarL, xL = sm.symbols(r'\overline{d^\prime}_L, '\
                       r'd^\prime_L', commutative=False)
Nc = xbarL * gamma * xL 
Nc = Nc.subs({xL: sm.cos(thetaC)*dL + sm.sin(thetaC)*sL,
              xbarL: sm.cos(thetaC)*dbarL + sm.sin(thetaC)*sbarL})
Nc = Nc.expand()
Nc
```

### The GIM Mechanism
In 1970, Sheldon Glashow, John
Iliopoulos and Luciano
Maiani (GIM) noted that if the strange quark were the lower component of another doublet, $$\begin{pmatrix} c_L \\ s_L \end{pmatrix},$$
whose upper component would be a new quark field, dubbed "charm", and if one generalized Cabibbo's idea and assumed that the lower components of the two quark doublets were mixed as follows
$$\begin{pmatrix} d_L^\prime \\ s_L^\prime \end{pmatrix} = 
\begin{pmatrix} \cos\theta_C & \sin\theta_C \\
-\sin\theta_C & \cos\theta_C \end{pmatrix} \begin{pmatrix} d_L \\  s_L \end{pmatrix},  $$
then the replacement
$$s_L \rightarrow s_L^\prime = -\sin\theta_C \, d_L + \cos\theta_C \, s_L$$
would yield exact cancellation of the unwanted FCNC interactions at the price of introducing yet another quark field, as demonstrated below.

### Implemention of GIM Mechanism

```python
ybarL, yL = sm.symbols(r'\overline{s^\prime}_L, '\
                                  r's^\prime_L', 
                                  commutative=False)
# add the GIM terms to the Cabbibo terms
yy = ybarL * gamma * yL
yy = yy.subs({yL:-sm.sin(thetaC)*dL + sm.cos(thetaC)*sL,
              ybarL:-sm.sin(thetaC)*dbarL + sm.cos(thetaC)*sbarL})
yy = yy.expand()
Nc = (Id3 - Qd * sm.sin(thetaW)**2) * Nc
Nc += (Id3 - Qd * sm.sin(thetaW)**2) * yy

# and simplify
Nc = Nc.simplify()
Nc
```

By introducing a 2nd doublet, the GIM mechanism neatly renders the weak neutral currents diagonal again, that is, it elimintes flavor-changing neutral currents and predicts the existence of a 4th quark, the charm quark. This prediction was confirmed in stunning fashion in 1974 with the indepedent discoveries of the $J/\psi$ at the Brookhaven National Laboratory (BNL) and the Stanford Linear Accelerator Center (SLAC).

Today, it is known that mixing of the down quark fields occurs across all three generations. The mixing is described by the Cabibbo-Kobayashi-Maskawa (CKM) matrix. Unfortunately, however, no one understands why mixing across the (three) generations should exist, nor why it takes the particular form that it does, nor what determines the strength of the mixing.

### Vector boson self-interactions
The pure vector boson Lagrangian is given by
\begin{align*}
    {\cal L}_{VB} & = - \frac{1}{4} \mathbf{W}_{\mu\nu} \cdot \mathbf{W}^{\mu\nu} - \frac{1}{4} B_{\mu\nu} B^{\mu\nu},
\end{align*}
where
\begin{align*}
    \mathbf{W}_{\mu\nu}
        & = \partial_\mu \mathbf{W}_\nu - \partial_\nu \mathbf{W}_\mu - g_2 \, \mathbf{W}_\mu \times \mathbf{W}_\nu, \\
    B_{\mu\nu} & = \partial_\mu B_\nu - \partial_\nu B_\mu.
\end{align*}

From
\begin{align}
    \begin{pmatrix} B_\mu \\ W^3_\mu \end{pmatrix} & = 
    \begin{pmatrix} 
        \cos\theta_W & -\sin\theta_W \\
        \sin\theta_W & \cos\theta_W
    \end{pmatrix}     \begin{pmatrix} A_\mu \\ Z_\mu \end{pmatrix} \\
    W^\pm_\mu & = \frac{1}{\sqrt{2}} (W^1_\mu \mp W^2_\mu)
\end{align}
we obtain
\begin{align*}
    W^1 & = (W^- + W^+)/\sqrt{2}\\
    W^2 & = (W^- - W^+)/\sqrt(2)\\
    W^3 & = \sin\theta_W A + \cos\theta_W Z\\
    B   & = \cos\theta_W A - \sin\theta_W Z,
\end{align*}
which we use below to rewrite the vector boson Lagrangian in terms of the usual fields.

```python
Wplusmu, Wminusmu = sm.symbols(r"W^+_\mu, W^-_\mu", commutative=False)
Wplusnu, Wminusnu = sm.symbols(r"W^+_\nu, W^-_\nu", commutative=False)

Zmu, Amu          = sm.symbols(r'Z_\mu, A_\mu',  commutative=False)
Znu, Anu          = sm.symbols(r'Z_\nu, A_\nu',  commutative=False)

dmu, dnu = sm.symbols(r'\partial_{\mu}, \partial_{\nu}', 
                      commutative=False)

W1mu, W2mu, W3mu  = sm.symbols(r'W^1_\mu W^2_\mu W^3_\mu',  
                               commutative=False)
W1nu, W2nu, W3nu  = sm.symbols(r'W^1_\nu W^2_\nu W^3_\nu',  
                               commutative=False)
Bmu, Bnu          = sm.symbols(r'B_\mu B_\nu',  commutative=False)

Vmu = Matrix([W1mu, W2mu, W3mu])
Vnu = Matrix([W1nu, W2nu, W3nu])

dnuVmu = dnu*Vmu
dmuVnu = dmu*Vnu
dnuBmu = dnu*Bmu
dmuBnu = dmu*Bnu

Vmu, Vnu, Bmu, Bnu, dnuVmu, dmuVnu, dnu*Bmu, dmu*Bnu
```

```python
VmuVnu  = Vmu.cross(Vnu)
VmuVnu
```

```python
WW  = dmuVnu - dnuVmu - q*VmuVnu / sm.sin(thetaW)
BB  = dmuBnu - dnuBmu
WW, BB
```

```python
Vsubs = {W1mu: (Wminusmu + Wplusmu)/sm.sqrt(2), 
         W2mu: (Wminusmu - Wplusmu)/sm.sqrt(2), 
         W3mu: sm.sin(thetaW)*Amu + sm.cos(thetaW)*Zmu, 
         W1nu: (Wminusnu + Wplusnu)/sm.sqrt(2), 
         W2nu: (Wminusnu - Wplusnu)/sm.sqrt(2),  
         W3nu: sm.sin(thetaW)*Anu + sm.cos(thetaW)*Znu,
         Bmu:  sm.cos(thetaW)*Amu - sm.sin(thetaW)*Zmu,
         Bnu:  sm.cos(thetaW)*Anu - sm.sin(thetaW)*Znu}
```

```python
Wplus_nu = sm.Symbol(r'W^+_{\nu}') 
Wplus_mu = sm.Symbol(r'W^+_{\mu}')

Wminus_nu = sm.Symbol(r'W^-_{\nu}') 
Wminus_mu = sm.Symbol(r'W^-_{\mu}')

A_nu = sm.Symbol(r'A_{\nu}') 
A_mu = sm.Symbol(r'A_{\mu}')

Z_nu = sm.Symbol(r'Z_{\nu}') 
Z_mu = sm.Symbol(r'Z_{\mu}') 

d_muWplus_nu = sm.Symbol(r'\partial_{\mu}W^+_{\nu}') 
d_nuWplus_mu = sm.Symbol(r'\partial_{\nu}W^+_{\mu}')

d_muWminus_nu = sm.Symbol(r'\partial_{\mu}W^-_{\nu}') 
d_nuWminus_mu = sm.Symbol(r'\partial_{\nu}W^-_{\mu}')

d_muA_nu = sm.Symbol(r'\partial_{\mu}A_{\nu}') 
d_nuA_mu = sm.Symbol(r'\partial_{\nu}A_{\mu}')

d_muZ_nu = sm.Symbol(r'\partial_{\mu}Z_{\nu}') 
d_nuZ_mu = sm.Symbol(r'\partial_{\nu}Z_{\mu}')

noncom = {Wplusnu: Wplus_nu, Wplusmu: Wplus_mu,
          Wminusnu: Wminus_nu, Wminusmu: Wminus_mu,
          Znu: Z_nu, Zmu: Z_mu, 
          Anu: A_nu, Amu: A_mu,
          dmu*Wplusnu: d_muWplus_nu, dnu*Wplusmu: d_nuWplus_mu,
          dmu*Wminusnu: d_muWminus_nu, dnu*Wminusmu: d_nuWminus_mu,
          dmu*Znu: d_muZ_nu, dnu*Zmu: d_nuZ_mu, 
          dmu*Anu: d_muA_nu, dnu*Amu: d_nuA_mu}
```

```python
for i in range(3):
    WW[i] = WW[i].subs(Vsubs).trigsimp().expand()
display_grouped(WW[0], terms_per_line=4)
```

```python
for i in range(3):
    WW[i] = WW[i].subs(noncom).expand()
display_grouped(WW[0], terms_per_line=4)
```

```python
BB = BB.subs(Vsubs).expand()
BB
```

```python
BB = BB.subs(noncom)
BB
```
